```python
data=pd.read_csv("Social_Network_Ads.csv")
data
```

```python
data=data.drop("User ID",axis=1)
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
dependent=data[["Purchased"]]
```

```python
independent=data[["Age", "EstimatedSalary","Gender_Male"]]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=10,weights='uniform')
classifier.fit(xtrain,ytrain)
```

```python
ypred=classifier.predict(xtest)
ypred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,ypred)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf=classification_report(ytest,ypred)
print(clf)
```
